This is a module that does image classification.

The run API is required by the interface for all modules. This is the method the server runs to start this module.

```python
def run(num_imgs):

  for idx in range (1, num_imgs + 1):
    img_path = '/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Input/' + str(idx) + '_input.jpg'
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    preds = model.predict(x)

    # TOP3 prediction
    # top3 = decode_predictions(preds, top=3)[0]
    # print(decode_predictions(preds, top=3)[0])
    # for _,classification,prob in top3:
    #   print(classification, prob)

    # TOP1 prediction
    predicted = decode_predictions(preds, top=3)[0][0][1]
    # Write strings to a file
    out_name = '/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Output/' + str(idx) + '_output.txt'
    out_txt = open(out_name,'w')
    out_txt.write(predicted)
    out_txt.close()
```
