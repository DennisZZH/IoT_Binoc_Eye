This is a module that turns the text in the image into a text file.

The run API is required by the interface for all modules. This is the method  the server runs to start this module.

```python
def run(num_imgs):
  for idx in range (1, num_imgs + 1):
    # extractedInformation = pytesseract.image_to_string(Image.open('test.png'))
    file_path = '/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Input/' + str(idx) + '_input.png'
    extractedInformation = pytesseract.image_to_string(Image.open(file_path))
    im = Image.open(file_path)

    # Write strings to a file
    out_name = '/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Output/' + str(idx) + '_output.txt'
    out_txt = open(out_name,'w')
    out_txt.write(extractedInformation)
    out_txt.close()

    
```
