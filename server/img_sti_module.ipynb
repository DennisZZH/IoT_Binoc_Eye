This ia a module that stitch images into a panoramic picture.

The run API is required by the interface for all modules. This is the method the server runs to start this module

```python
def run(num_imgs) :
  imgs = []
  for idx in range (1, num_imgs + 1):
    img = cv2.imread('/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Input/' + str(idx) + '_input.JPG')
    imgs.append(img)
  
  stitcher = cv2.Stitcher.create()
  status, pano = stitcher.stitch(imgs)
  if status != cv2.Stitcher_OK:
    print("Can't stitch images, error code = %d" % status)
    sys.exit(-1)
  
  cv2.imwrite('/content/drive/Shareddrives/Bionic_Eye_IoT_Data/Output/0_output.jpeg', pano)
  print('img_sti_module complete!')
```
